# DATA CLEANING

checking for missing values

```python
missing_values = data.isnull().sum()
missing_values
```

From the data the following columns have missing values:chol, hdl,glyhb, height,weight,frame,bp.1s,bp.1d,bp.2s,bp.2d,waist,hip & time.ppn. bp.2s & bp.2d columns have a high number of missing values (262 for each), so we shall drop the columns.id, frame, location & time as .ppn doesn't have a significant impact on your analysis & doesn't provide any useful information so we shall also drop them

```python
#Dropping unnecessary columns
columns_to_delete = ['id', 'location','bp.2s','bp.2d', 'time.ppn', 'frame']
data.drop(columns = columns_to_delete, inplace = True)

df = data
```

Handling missing values

```python
#Identifying numerical columns with missing values
numerical_cols = df.select_dtypes(include=['number']).columns

# Impute missing values in numerical columns with the mean
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].mean())
```

```python
df.info()
```

Our data now has no missing values.

```python
#rename columns
column_name = {
    'chol': 'Cholesterol',
    'stab.glu': 'Glucose stability',
    'hdl': 'High-Density Lipoprotein',
    'ratio': 'Cholesterol ratio',
    'glyhb': 'Glycosylated hemoglobin',
    'gender': 'Gender',
    'height': 'Height',
    'weight': 'Weight',
    'bp.1s' : 'Systolic blood pressure',
    'bp.1d' : 'Diastolic blood pressure',
    'waist' : 'Waist circumference',
    'hip' :'Hip circumference'
}
diabetes = df.rename(columns=column_name)
```

```python
diabetes.head()
```

```python
diabetes['diabetes_status'] = diabetes['Glycosylated hemoglobin'].apply(lambda a: 'no_diabetes' if a < 5.7 else ('pre_diabetes' if 5.7 <= a < 6.4 else 'diabetic'))
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
diabetes['diabetes_status_encoded'] = le.fit_transform(diabetes['diabetes_status'])
le.fit(diabetes['Gender'])
diabetes['Gender'] = le.transform(diabetes['Gender'])
```

```python
#removing the non-numeric column
diabetes = diabetes.drop(columns=['diabetes_status'])
```
